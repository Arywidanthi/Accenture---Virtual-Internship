```python
import pandas as pd
import numpy as np
```

# Content

```python
content = pd.read_csv('Content.csv')
content
```

delete columns

```python
# Check duplicate data
duplicates = content.duplicated()
duplicates.sum()
```

```python
content1 = content.drop(['Unnamed: 0', 'User ID','URL'], axis=1)
```

```python
content1
```

# Reactions

```python
reactions = pd.read_csv('Reactions.csv')
reactions
```

```python
# Check duplicate data
duplicates = reactions.duplicated()
duplicates.sum()
```

```python
reactions.isna().sum()
```

```python
reactions1 = reactions.drop(['Unnamed: 0', 'User ID', 'Datetime'], axis=1)
```

```python
reactions1
```

```python
reactions1_missing = reactions1[reactions['Type'].isna()]
reactions1_missing
```

```python
reactions1 = reactions1.drop(reactions1_missing.index)
```

```python
reactions1
```

# ReactionTypes

```python
reactiontypes = pd.read_csv('ReactionTypes.csv')
reactiontypes
```

```python
duplicates = reactiontypes.duplicated()
duplicates.sum()
```

```python
reactiontypes.rename(columns={'Type': 'Reaction Type'}, inplace=True)
```

```python
reactiontypes
```

# Merge

```python
combined_table = pd.merge(content1, reactions1, on=['Content ID'], how='inner')
```

```python
combined_table
```

```python
combined_table1 = combined_table.rename(columns={'Type_x': 'Content Type', 'Type_y': 'Reaction Type'})
```

```python
combined_table1
```

```python
combined_table_final = pd.merge(combined_table1, reactiontypes, on=['Reaction Type'], how='inner')
```

```python
combined_table_final
```

```python
combined_table_final.to_excel('Cleaned Table.xlsx', index=False)
```
